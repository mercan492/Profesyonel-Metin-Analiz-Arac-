```python
import string
from collections import Counter # Sayma işlemleri için profesyonel araç

def kelime_frekansi_analizi(metin):
    if not metin or metin.strip() == "":
        return None

    # 1. Metni temizle (Noktalama işaretlerini kaldır ve küçük harfe çevir)
    tablo = str.maketrans('', '', string.punctuation)
    temiz_metin = metin.translate(tablo).lower()
    
    # 2. Kelimeleri ayır
    kelimeler = temiz_metin.split()
    
    # 3. Kelime Sayılarını Hesapla (En önemli kısım!)
    # Counter, listedeki her bir kelimenin kaç kez geçtiğini otomatik sayar
    kelime_sayaclari = Counter(kelimeler)
    
    # En çok kullanılan ilk 3 kelimeyi al
    en_cok_gecenler = kelime_sayaclari.most_common(3)
    
    # 4. Raporu hazırla
    analiz = {
        "toplam_kelime": len(kelimeler),
        "benzersiz_kelime": len(kelime_sayaclari),
        "en_populer_kelimeler": en_cok_gecenler,
        "en_uzun_kelime": max(kelimeler, key=len) if kelimeler else ""
    }
    
    return analiz

# --- Uygulama Kısmı ---
metin_input = input("Analiz edilecek metni veya paragrafı buraya yapıştırın:\n")
rapor = kelime_frekansi_analizi(metin_input)

if rapor:
    print("\n" + "*" * 20)
    print("   PROFESYONEL METİN ANALİZİ   ")
    print("*" * 20)
    print(f" Toplam Kelime Sayısı    : {rapor['toplam_kelime']}")
    print(f" Benzersiz Kelime Sayısı : {rapor['benzersiz_kelime']}")
    print(f" En Uzun Kelime          : '{rapor['en_uzun_kelime']}'")
    
    print("\n En Çok Kullanılan Kelimeler:")
    for kelime, sayi in rapor['en_populer_kelimeler']:
        # Her kelime için küçük bir çubuk grafik simülasyonu
        grafik = "■" * sayi
        print(f"  - {kelime.capitalize()}: {sayi} kez {grafik}")
    
    print("\n" + "=" * 40)
else:
    print(" Analiz için geçerli bir metin bulunamadı.")
```
